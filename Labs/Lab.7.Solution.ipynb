Write a function that creates pair plots and use it to compare variables in the SUSY sample, separately for low and high-level features. Refer to Lecture 13 for details. Do not use `seaborn`.


```python
def pair_plot(df, variables, n=5000):
    sample = df.sample(n)
    for i, var1 in enumerate(variables):
        for j, var2 in enumerate(variables):
            if j <= i: continue
            plt.figure(figsize=(6, 4))
            plt.scatter(sample[var1], sample[var2], alpha=0.1, s=1)
            plt.xlabel(var1)
            plt.ylabel(var2)
            plt.title(f"{var1} vs {var2}")
            plt.grid(True)
            plt.show()
```

```python
from IPython.display import HTML, display
import tabulate

def show_corr_cov_matrices(df, columns, name):
    cov_matrix = np.cov(df[columns].T)
    corr_matrix = np.corrcoef(df[columns].T)
    headers = [""] + columns
    cov_table = [[columns[i]] + list(np.round(row, 3)) for i, row in enumerate(cov_matrix)]
    corr_table = [[columns[i]] + list(np.round(row, 3)) for i, row in enumerate(corr_matrix)]
    display(HTML(f"<h4>{name} Covariance Matrix</h4>"))
    display(HTML(tabulate.tabulate(cov_table, tablefmt='html', headers=headers)))
    display(HTML(f"<h4>{name} Correlation Matrix</h4>"))
    display(HTML(tabulate.tabulate(corr_table, tablefmt='html', headers=headers)))
```

### Exercise 5.2

Plot the True Positive Rate (TPR) (aka signal efficiency $\epsilon_S(x_c)$) and False Positive Rate (FPR) (aka background efficiency $\epsilon_B(x_c)$) as function of $x_c$ for applying the strategy in part a to each observable. 


```python
def plot_efficiencies(df, var):
    thresholds = np.linspace(df[var].min(), df[var].max(), 100)
    eps_s, eps_b = [], []
    for x_c in thresholds:
        eps_s.append((df_sig[var] > x_c).sum() / len(df_sig))
        eps_b.append((df_bkg[var] > x_c).sum() / len(df_bkg))
    plt.plot(thresholds, eps_s, label="Signal efficiency")
    plt.plot(thresholds, eps_b, label="Background efficiency")
    plt.title(f"Efficiencies vs {var}")
    plt.xlabel(var)
    plt.ylabel("Efficiency")
    plt.legend()
    plt.grid()
    plt.show()
```

### Exercise 5.3
Assume 3 different scenarios corresponding to different numbers of signal and background events expected in data:

1. Expect $N_S=10$, $N_B=100$.
1. Expect $N_S=100$, $N_B=1000$.
1. Expect $N_S=1000$, $N_B=10000$.
1. Expect $N_S=10000$, $N_B=100000$.

Plot the significance ($\sigma_{S'}$) for each observable as function of $x_c$ for each scenario, where 

$\sigma_{S'}= \frac{N'_S}{\sqrt{N'_S+N'_B}}$

and $N'_{S,B} = \epsilon_{S,B}(x_c) * N_{S,B}$.

```python
def significance_plot(df, var, scenarios):
    thresholds = np.linspace(df[var].min(), df[var].max(), 100)
    for Ns, Nb in scenarios:
        sig = []
        for x_c in thresholds:
            eps_s = (df_sig[var] > x_c).sum() / len(df_sig)
            eps_b = (df_bkg[var] > x_c).sum() / len(df_bkg)
            Ns_p = eps_s * Ns
            Nb_p = eps_b * Nb
            sig.append(Ns_p / np.sqrt(Ns_p + Nb_p + 1e-6))
        plt.plot(thresholds, sig, label=f"Ns={Ns}, Nb={Nb}")
    plt.title(f"Significance vs {var}")
    plt.xlabel(var)
    plt.ylabel("σ_s")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
scenarios = [(10, 100), (100, 1000), (1000, 10000), (10000, 100000)]
significance_plot(df, "MT2", scenarios)
```

### Exercise 6.3
In 3c above you computed the significance for each observable assuming to make no other selections on any other observable. If the variables are correlated, then this assumption can lead to non-optimial results when selecting on multiple variables. By looking at the correlation matrices and your answers to 4b, identify where this effect could be most detrimental to the significance. Attempt to correct the issue by applying the selection in one observable and then optimizing (part 3c) for a second observable. What happens if you change the order of your selection (make selection on second and optimize on first)?




```python
def cut_flow(df, var_cuts, Ns, Nb):
    df_cut = df.copy()
    cuts = []
    for var, direction, x_c in var_cuts:
        if direction == '>':
            df_cut = df_cut[df_cut[var] > x_c]
        elif direction == '<':
            df_cut = df_cut[df_cut[var] < x_c]
        s_eff = len(df_cut[df_cut.signal == 1]) / Ns
        b_eff = len(df_cut[df_cut.signal == 0]) / Nb
        Ns_eff = s_eff * Ns
        Nb_eff = b_eff * Nb
        sigma = Ns_eff / np.sqrt(Ns_eff + Nb_eff + 1e-6)
        cuts.append([var, direction + str(x_c), f"{s_eff:.3f}", f"{b_eff:.3f}", int(Ns_eff), int(Nb_eff), f"{sigma:.2f}"])
    return cuts

# Example usage
cut_flow_table = cut_flow(df, [("MT2", ">", 0.5), ("M_R", ">", 1.0), ("cos_theta_r1", "<", 0.5)], 10000, 100000)

from IPython.display import HTML
import tabulate
display(HTML(tabulate.tabulate(cut_flow_table, headers=["Var", "Cut", "ε_S", "ε_B", "N'_S", "N'_B", "σ_S"], tablefmt="html")))
```

## Exercise 8: Linear Discriminant

### Exercise 8.1

Using numpy, compute the between-class $\bf{S}_B$ and within-class $\bf{S}_W$ covariance matrices defined as:

$$
\bf{S}_B = (\bf{m_2}-\bf{m_1})(\bf{m_2}-\bf{m_1})^T \\
$$
$$
\bf{S}_W = \sum_{i=1,2} \sum_{n=1}^{l_i} (\bf{x}_n^i - \bf{m}_i) (\bf{x}_n^i - \bf{m}_i)^T
$$

where $\bf{m_i}$ are the vectors containing the means for category 1 and 2, here defined as signal and background. Here $\bf{x}_n^i$ is the vector containing the observables for the $n$th example event in category $i$.


```python
# Mean vectors
m1 = df_bkg[VarNames[1:]].mean().values
m2 = df_sig[VarNames[1:]].mean().values

# Covariance matrices
SW = np.cov(df_bkg[VarNames[1:]].values, rowvar=False) + np.cov(df_sig[VarNames[1:]].values, rowvar=False)
SB = np.outer((m2 - m1), (m2 - m1))

# Linear discriminant coefficients
w = np.linalg.inv(SW).dot(m2 - m1)
```

### Exercise 8.1.2

Compute the linear coefficients $\bf{w} = \bf{S_W}^{-1}(\bf{m_2}-\bf{m_1})$. Compare the histogram of the distribution of $F_n^i=\bf{w}^T\bf{x}_n^i$ for the two categories.


```python
# Project signal and background
Fn_sig = df_sig[VarNames[1:]].dot(w)
Fn_bkg = df_bkg[VarNames[1:]].dot(w)

plt.figure(figsize=(10,5))
plt.hist(Fn_sig, bins=100, alpha=0.5, label="Signal", density=True)
plt.hist(Fn_bkg, bins=100, alpha=0.5, label="Background", density=True)
plt.legend()
plt.title("Linear Discriminant Projection")
plt.xlabel("F_n")
plt.grid()
plt.show()
```

### Exercise 8.1.3

Draw the ROC curve for $F_n$. 


```python
Fn = df[VarNames[1:]].dot(w)
fpr, tpr, _ = roc_curve(df["signal"], Fn)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f"LDA ROC (AUC = {roc_auc:.2f})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for LDA")
plt.legend()
plt.grid()
plt.show()
```
