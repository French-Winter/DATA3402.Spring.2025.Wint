1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self,maximum,minimum=0,count=0):
        self.maximum=maximum
        self.minimum=minimum
        self.count=count
        
    # Module to reset counter
    def reset(self):
        if self.count !=0:
            self.count=0
            
    # Module to add to counter by a given increment <= maximum
    def incr(self,increment):
        original_count=[]
        original_count.append(self.count)
        self.count +=increment
        if self.count>self.maximum:
            self.count=original_count[0]
            print(f"Error: Counting with increment {increment} surpasses the maximum.")
            print("Try again with a smaller increment value.")
            return self.count
        else:
            return self.count

# Initialize counter
dracula=Counter(10)

# Increment counter
dracula.incr(2)

# Check if error when over maximum
dracula.incr(11)
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
# Import math dictionary to use in RightTriangle class
import math

class Shape:
    def __init__(self,x_coor,y_coor,max_pairs=16):
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs

    def show_x_y(self):
        return (self.x_coor,self.y_coor)

    def area(self):
        pass 

    # Module to show an error message if the number of points inputted is past the maximum
    def base_points(self,number):
        if abs(number)>self.__max_pairs:
            print(f"Error: Number of points '{number}' has exceeded points"
            f" limit '{self.__max_pairs}'.")
            print(f"Try again with number of points less than or equal to {self.__max_pairs}.")
    
class Rectangle(Shape):
    def __init__(self,length,width,x_coor,y_coor,max_pairs=16):
        super(Rectangle,self).__init__(x_coor,y_coor,max_pairs)
        self.__length=length
        self.__width=width
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs

    def show_corner(self):
        return Shape.show_x_y(self)

    def show_length(self):
        return self.__length

    def show_width(self):
        return self.__width

    def area(self):
        area=self.__length*self.__width
        return area

    def perimeter(self):
        perimeter=2*(self.__length+self.__width)
        return perimeter

    # Module to divide a given line in the object into a given number of points
    def points(self,number,axis):
        if abs(number)<=self.__max_pairs:  # checks if number of points is applicable
            if axis=='Y_min' or axis=='y_min':  # divides the left Y axis
                range_y=self.__width
                increment_y=range_y/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.__x_coor,y))  # x coordinate stays the same, and is the right y axis
                return points
            elif axis=='X_min' or axis=='x_min':  # divides the lower X axis
                range_x=self.__length
                increment_x=range_x/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                x=self.__x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.__y_coor))  # y coordinate stays the same, and is lower x axis
                return points
            elif axis=='Y_max' or axis=='y_max':  # divides the right Y axis
                range_y=self.__width
                increment_y=range_y/abs(number)
                points=[(self.__x_coor+self.__length,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.__x_coor+self.__length,y))  # x coordinate stays the same, and is the left y axis
                return points
            elif axis=='X_max' or axis=='x_max':  # divides the top X axis
                range_x=self.__length
                increment_x=range_x/abs(number)
                points=[(self.__x_coor,self.__y_coor+self.__width)]
                x=self.__x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.__y_coor+self.__width))  # y coordinate stays the same, and is the top x axis
                return points
        else:
              return Shape.base_points(self,number)  # if number of points surpasses maximum, module inherits from parent
                                                    # class and returns error message

class Circle(Shape):
    def __init__(self,radius,x_coor,y_coor,max_pairs=16):
        super().__init__(x_coor,y_coor,max_pairs)
        self.__radius=radius
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs

    def show_centre(self):
        return Shape.show_x_y(self)

    def show_radius(self):
        return self.__radius

    def area(self):
        return (22/7)*(self.__radius**2)

    def circumference(self):
        return 2*(22/7)*self.__radius

    def points(self,number,angle):
        if abs(number)<=self.__max_pairs:
            if angle==90:  # tangent function which does not exist at 90°
                # below is for cases equivalent to tan(90)
                range_y=self.__radius
                increment_y=range_y/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.__x_coor,y))
                return points
            elif angle==270:
                # below is for cases equivalent to tan(270)
                range_y=self.__radius
                increment_y=range_y/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y -=increment_y
                    points.append((self.__x_coor,y))
                return points
            else:  # for all other angles 
                if angle>=0 and angle<90:  # all positive x and y values for first quadrant
                    angle_radians=math.radians(angle)
                    range_y=math.tan(angle_radians)*self.__radius
                    increment_y=range_y/abs(number)
                    range_x=range_y/math.tan(angle_radians)
                    increment_x=range_x/abs(number)
                    points=[(self.__x_coor,self.__y_coor)]
                    x=self.__x_coor
                    y=self.__y_coor
                    for i in range(abs(number)):
                        x -=increment_x
                        y -=increment_y
                        points.append((x,y))
                    return points
                elif angle>90 and angle<180:  # positive y values and negative x values for second quadrant
                    angle_radians=math.radians(angle)
                    range_y=math.tan(angle_radians)*self.__radius
                    increment_y=range_y/abs(number)
                    range_x=range_y/math.tan(angle_radians)
                    increment_x=range_x/abs(number)
                    points=[(self.__x_coor,self.__y_coor)]
                    x=self.__x_coor
                    y=self.__y_coor
                    for i in range(abs(number)):
                        x -=increment_x
                        y +=increment_y
                        points.append((x,y))
                    return points
                elif angle>=180 and angle<270:  # all negative x and y values for third quadrant
                    angle_radians=math.radians(angle)
                    range_y=math.tan(angle_radians)*self.__radius
                    increment_y=range_y/abs(number)
                    range_x=range_y/math.tan(angle_radians)
                    increment_x=range_x/abs(number)
                    points=[(self.__x_coor,self.__y_coor)]
                    x=self.__x_coor
                    y=self.__y_coor
                    for i in range(abs(number)):
                        x -=increment_x
                        y -=increment_y
                        points.append((x,y))
                    return points
                elif angle>270 and angle<360:  # positive x values and negative y values for fourth quadrant
                    angle_radians=math.radians(angle)
                    range_y=math.tan(angle_radians)*self.__radius
                    increment_y=range_y/abs(number)
                    range_x=range_y/math.tan(angle_radians)
                    increment_x=range_x/abs(number)
                    points=[(self.__x_coor,self.__y_coor)]
                    x=self.__x_coor
                    y=self.__y_coor
                    for i in range(abs(number)):
                        x +=increment_x
                        y -=increment_y
                        points.append((x,y))
                    return points
        else:
            return Shape.base_points(self,number)

class RightTriangle(Shape):
    def __init__(self,base,height,x_coor,y_coor,max_pairs=16):
        super().__init__(x_coor,y_coor,max_pairs)
        self.__base=base
        self.__height=height
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs

    def show_corner(self):
        return Shape.show_x_y(self)

    def show_base(self):
        return self.__base

    def show_height(self):
        return self.__height

    def area(self):
        return 0.5*(self.__base*self.__height)

    def perimeter(self):
        hypotenuse=(self.__base**2+self.__height**2)**0.5
        return hypotenuse +self.__base +self.__height

    def points(self,number,axis):
        if abs(number)<=self.__max_pairs:
            if axis=='Y' or axis=='y':  # divides the height axis of the right triangle
                range_y=self.__height
                increment_y=range_y/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.__x_coor,y))
                return points
            elif axis=='X' or axis=='x':  # divides the base axis of the right triangle
                range_x=self.__base
                increment_x=range_x/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                x=self.__x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.__y_coor))
                return points
            elif axis=='XY' or axis=='xy':  # divides the hypotenuse of the right triangle
                range_x=abs(self.__base)
                increment_x=range_x/abs(number)
                range_y=abs(self.__height)
                increment_y=range_y/abs(number)
                points=[(self.__x_coor+self.__base,self.__y_coor)]
                x=self.__x_coor+self.__base
                y=self.__y_coor
                if self.__height>0:
                    if self.__base<0:
                        for i in range(abs(number)-1):
                            x +=increment_x
                            y +=increment_y
                            points.append((x,y))
                    elif self.__base>0:
                        for i in range(abs(number)-1):
                            x -=increment_x
                            y +=increment_y
                            points.append((x,y))  
                if self.__height<0:
                    if self.__base<0:
                        for i in range(abs(number)-1):
                            x +=increment_x
                            y -=increment_y
                            points.append((x,y))
                    elif self.__base>0:
                        for i in range(abs(number)-1):
                            x -=increment_x
                            y -=increment_y
                            points.append((x,y))
                points.append((self.__x_coor,self.__y_coor+self.__height))
                return points
        else:
            return Shape.base_points(self,number)
    
# Initialize objects
box=Rectangle(3,2,3,2)
circ=Circle(3,3,3)
right=RightTriangle(-3,4,0,0)

# Check if all verisons of 'points' module is working
print("Bottom X:",box.points(5,'x_min'))
print("\nLeft Y:",box.points(5,'y_min'))
print("\nTop X:",box.points(5,'x_max'))
print("\nRight Y:",box.points(5,'y_max'))
print("\nHorizontal Points:",circ.points(5,180))
print("\nVertical Points:",circ.points(5,90))
print("\nDiagonal Points:",circ.points(5,45))
print("\nBase Points:",right.points(5,'x'))
print("\nHeight Points:",right.points(5,'y'))
print("\nHypotenuse Points:",right.points(5,'xy'))
```

```python
class Shape:
    def __init__(self,x_coor,y_coor,max_pairs=16):
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs

    def show_x_y(self):
        return (self.x_coor,self.y_coor)

    def area(self):
        pass 

    def base_points(self,number):
        if abs(number)>self.__max_pairs:
            print(f"Error: Number of points '{number}' has exceeded points"
            f" limit '{self.__max_pairs}'.")
            print(f"Try again with number of points less than or equal to {self.__max_pairs}.")
    
class Rectangle(Shape):
    def __init__(self,length,width,x_coor,y_coor,max_pairs=16):
        super(Rectangle,self).__init__(x_coor,y_coor,max_pairs)
        self.__length=length
        self.__width=width
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs

    def show_corner(self):
        return Shape.show_x_y(self)

    def show_length(self):
        return self.__length

    def show_width(self):
        return self.__width

    def area(self):
        area=self.__length*self.__width
        return area

    def perimeter(self):
        perimeter=2*(self.__length+self.__width)
        return perimeter

    def points(self,number,axis):
        if abs(number)<=self.__max_pairs:
            if axis=='Y_min' or axis=='y_min':
                range_y=self.__width
                increment_y=range_y/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.__x_coor,y))
                return points
            elif axis=='X_min' or axis=='x_min':
                range_x=self.__length
                increment_x=range_x/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                x=self.__x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.__y_coor))
                return points
            elif axis=='Y_max' or axis=='y_max':
                range_y=self.__width
                increment_y=range_y/abs(number)
                points=[(self.__x_coor+self.__length,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.__x_coor+self.__length,y))
                return points
            elif axis=='X_max' or axis=='x_max':
                range_x=self.__length
                increment_x=range_x/abs(number)
                points=[(self.__x_coor,self.__y_coor+self.__width)]
                x=self.__x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.__y_coor+self.__width))
                return points
        else:
              return Shape.base_points(self,number)

    # Module to show if a given point is inside the bounds of the object
    def x_y_inside(self,x,y):
        return (self.__x_coor<=x<=(self.__length+self.__x_coor)  # checks if the x is in the interval of the x axis
        and self.__y_coor<=y<=(self.__width+self.__y_coor))  # checks if the y is in the interval of the y axis  
            
class Circle(Shape):
    def __init__(self,radius,x_coor,y_coor,max_pairs=16):
        super().__init__(x_coor,y_coor,max_pairs)
        self.__radius=radius
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs

    def show_centre(self):
        return Shape.show_x_y(self)

    def show_radius(self):
        return self.__radius

    def area(self):
        return (22/7)*(self.__radius**2)

    def circumference(self):
        return 2*(22/7)*self.__radius

    def points(self,number,angle):
        if abs(number)<=self.__max_pairs:
            if (not angle==0 and angle%90==0) and not angle%180==0:
                range_y=self.__radius
                increment_y=range_y/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.__x_coor,y))
                return points
            else:
                if angle>90 and angle<180:
                    angle_radians=math.radians(angle)
                    range_y=math.tan(angle_radians)*self.__radius
                    increment_y=range_y/abs(number)
                    range_x=range_y/math.tan(angle_radians)
                    increment_x=range_x/abs(number)
                    points=[(self.__x_coor,self.__y_coor)]
                    x=self.__x_coor
                    y=self.__y_coor
                    for i in range(abs(number)):
                        x -=increment_x
                        y +=increment_y
                        points.append((x,y))
                    return points
                elif angle>=180 and angle<270:
                    angle_radians=math.radians(angle)
                    range_y=math.tan(angle_radians)*self.__radius
                    increment_y=range_y/abs(number)
                    range_x=range_y/math.tan(angle_radians)
                    increment_x=range_x/abs(number)
                    points=[(self.__x_coor,self.__y_coor)]
                    x=self.__x_coor
                    y=self.__y_coor
                    for i in range(abs(number)):
                        x -=increment_x
                        y -=increment_y
                        points.append((x,y))
                    return points
                elif angle>270 and angle<360:
                    angle_radians=math.radians(angle)
                    range_y=math.tan(angle_radians)*self.__radius
                    increment_y=range_y/abs(number)
                    range_x=range_y/math.tan(angle_radians)
                    increment_x=range_x/abs(number)
                    points=[(self.__x_coor,self.__y_coor)]
                    x=self.__x_coor
                    y=self.__y_coor
                    for i in range(abs(number)):
                        x +=increment_x
                        y -=increment_y
                        points.append((x,y))
                    return points
        else:
            return Shape.base_points(self,number)

    # Module will check if the distance of the point from the circle centre is greater than the radius
    def x_y_inside(self,x,y):  
        return (abs(x-self.__x_coor)<=self.__radius  
                and abs(y-self.__y_coor)<=self.__radius)
        
class RightTriangle(Shape):
    def __init__(self,base,height,x_coor,y_coor,max_pairs=16):
        super().__init__(x_coor,y_coor,max_pairs)
        self.__base=base
        self.__height=height
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs
        self.__x2=self.__x_coor
        self.__y2=self.__y_coor+self.__height
        self.__x3=self.__x_coor+self.__base
        self.__y3=self.__y_coor
        
    def show_corner(self):
        return Shape.show_x_y(self)

    def show_base(self):
        return self.__base

    def show_height(self):
        return self.__height

    # Adjust the area module to be used in the x_y_inside module
    def area(self,x1=None,y1=None,x2=None,y2=None,x3=None,y3=None):
        if x1==None:
            x1=self.__x_coor
        if y1==None:
            y1=self.__y_coor
        if x2==None:
            x2=self.__x2
        if y2==None:
            y2=self.__y2
        if x3==None:
            x3=self.__x3
        if y3==None:
            y3=self.__y3
        return abs(x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2))/2

    def perimeter(self):
        hypotenuse=(self.__base**2+self.__height**2)**0.5
        return hypotenuse +self.__base +self.__height

    def points(self,number,axis):
        if abs(number)<=self.__max_pairs:
            if axis=='Y' or axis=='y':
                range_y=self.__height
                increment_y=range_y/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                y=self.__y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.__x_coor,y))
                return points
            elif axis=='X' or axis=='x':
                range_x=self.__base
                increment_x=range_x/abs(number)
                points=[(self.__x_coor,self.__y_coor)]
                x=self.__x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.__y_coor))
                return points
            elif axis=='XY' or axis=='xy':
                range_x=abs(self.__base)
                increment_x=range_x/abs(number)
                range_y=abs(self.__height)
                increment_y=range_y/abs(number)
                points=[(self.__x_coor+self.__base,self.__y_coor)]
                x=self.__x_coor+self.__base
                y=self.__y_coor
                if self.__height>0:
                    if self.__base<0:
                        for i in range(abs(number)-1):
                            x +=increment_x
                            y +=increment_y
                            points.append((x,y))
                    elif self.__base>0:
                        for i in range(abs(number)-1):
                            x -=increment_x
                            y +=increment_y
                            points.append((x,y))  
                if self.__height<0:
                    if self.__base<0:
                        for i in range(abs(number)-1):
                            x +=increment_x
                            y -=increment_y
                            points.append((x,y))
                    elif self.__base>0:
                        for i in range(abs(number)-1):
                            x -=increment_x
                            y -=increment_y
                            points.append((x,y))
                points.append((self.__x_coor,self.__y_coor+self.__height))
                return points
        else:
            return Shape.base_points(self,number)
    
    # Module will make three new triangles using the given point
    # if the areas of the three triangles is equal to the area of the triangle object
    # then yes the point is inside the triangle
    def x_y_inside(self,x,y):
        area=self.area()
        area1=self.area(x1=x,y1=y)
        area2=self.area(x2=x,y2=y)
        area3=self.area(x3=x,y3=y)
        return area==area1 + area2 + area3
    
# Initialize objects
box=Rectangle(3,2,3,2)
circ=Circle(3,3,3)
right=RightTriangle(-3,4,7,5)

# Check if random points are inside
print("(3,5) Inside Rectangle:",box.x_y_inside(3,5))
print("(3,5) Inside Circle:",circ.x_y_inside(3,5))
print("(3,5) Inside RightTriangle:",right.x_y_inside(3,5))
```

```python
class Shape:
    def __init__(self,x_coor,y_coor,max_pairs=16):
        self.__x_coor=x_coor
        self.__y_coor=y_coor
        self.__max_pairs=max_pairs

    def show_x_y(self):
        return (self.__x_coor,self.__y_coor)

    def base_points(self,number):
        if abs(number)>self.__max_pairs:
            print(f"Error: Number of points '{number}' has exceeded points"
            f" limit '{self.__max_pairs}'.")
            print(f"Try again with number of points less than or equal to {self.__max_pairs}.")

    # Module will take 2 objects and check with they are within the bounds of each other
    # If one of the objects is a circle, input the circle first
    # If both objects are circles, the axis will be None
    def check_overlap(self,shape1,shape2,axis):
        if axis==None:  # if both objects are circles
            x1,y1=shape1.show_centre()
            x2,y2=shape2.show_centre()
            dist=shape1.show_radius()
            return (abs(x1-x2)<=dist and abs(y1-y2)<=dist)
        else:  # if one or both of the objects have sides
            checks=[]
            for i in axis:
                lst=shape2.points(16,i)
                for j in lst:
                    x,y=j
                    checks.append(shape1.x_y_inside(x,y))
            if True in checks: 
                return True  # returns True if any of the points in second object is in the first object
            else:
                return False  # returns False if none of the points in the second object is in the first object
            
            
# Initialize Shape object
shape=Shape(0,0)

# Check if 'check_overlap' module is working
print("Circles overlap:",shape.check_overlap(circ,circ,None))
print("Circle and Rectangle overlap:",shape.check_overlap(circ,box,['x_min','y_min','x_max','y_max']))
print("Rectangle and Right angle Triangle overlap:",shape.check_overlap(box,right,['x','y','xy']))
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]
        self.shapes=[]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]

    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]

    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)

    def line(self, x1, y1, x2, y2, **kargs):
        slope=(y2-y1) / (x2-x1)
        intercept=y1-(x1*slope)
        for y in range(y1,y2):
            x=int(slope*y + intercept)
            self.set_pixel(x,y, **kargs)

    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
        

## I will change the attributes of the objects to not private
## For the sake of the upcoming Raster Drawing class
## I will also add an attribute 'name' for the upcoming Raster Drawing class
class Shape:
    def __init__(self,x_coor,y_coor,max_pairs=16,**kwargs):
        self.x_coor=x_coor
        self.y_coor=y_coor
        self.__max_pairs=max_pairs
        self.kwargs=kwargs

    def show_x_y(self):
        return (self.x_coor,self.y_coor)

    def base_points(self,number):
        if abs(number)>self.__max_pairs:
            print(f"Error: Number of points '{number}' has exceeded points"
            f" limit '{self.__max_pairs}'.")
            print(f"Try again with number of points less than or equal to {self.__max_pairs}.")

    def check_overlap(self,shape1,shape2,axis):
        if axis==None:
            x1,y1=shape1.show_centre()
            x2,y2=shape2.show_centre()
            dist=shape1.show_radius()
            return (abs(x1-x2)<=dist and abs(y1-y2)<=dist)
        else:
            checks=[]
            for i in axis:
                lst=shape2.points(16,i)
                for j in lst:
                    x,y=j
                    checks.append(shape1.x_y_inside(x,y))
            if True in checks:
                return True
            else:
                return False
            
class Rectangle(Shape):
    def __init__(self,length,width,x_coor,y_coor,max_pairs=16,name="",**kwargs):
        super(Rectangle,self).__init__(x_coor,y_coor,max_pairs)
        self.length=length
        self.width=width
        self.x_coor=x_coor
        self.y_coor=y_coor
        self.__max_pairs=max_pairs
        self.name=name
        self.kwargs=kwargs

    def name(self):
        return self.name
    
    def show_corner(self):
        return Shape.show_x_y(self)

    def show_length(self):
        return self.length

    def show_width(self):
        return self.width

    def area(self):
        area=self.length*self.width
        return area

    def perimeter(self):
        perimeter=2*(self.length+self.width)
        return perimeter

    def points(self,number,axis):
        if abs(number)<=self.__max_pairs:
            if axis=='Y_min' or axis=='y_min':
                range_y=self.width
                increment_y=range_y/abs(number)
                points=[(self.x_coor,self.y_coor)]
                y=self.y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.x_coor,y))
                return points
            elif axis=='X_min' or axis=='x_min':
                range_x=self.length
                increment_x=range_x/abs(number)
                points=[(self.x_coor,self.y_coor)]
                x=self.x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.y_coor))
                return points
            elif axis=='Y_max' or axis=='y_max':
                range_y=self.width
                increment_y=range_y/abs(number)
                points=[(self.x_coor+self.length,self.y_coor)]
                y=self.y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.x_coor+self.length,y))
                return points
            elif axis=='X_max' or axis=='x_max':
                range_x=self.length
                increment_x=range_x/abs(number)
                points=[(self.x_coor,self.y_coor+self.width)]
                x=self.x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.y_coor+self.width))
                return points
        else:
              return Shape.base_points(self,number)

    def x_y_inside(self,x,y):
        return (self.x_coor<=x<=(self.length+self.x_coor)
        and self.y_coor<=y<=(self.width+self.y_coor))       
  
    def paint(self,canvas):
        canvas.v_line(self.x_coor,self.y_coor,self.width, **self.kwargs)
        canvas.v_line(self.x_coor,self.y_coor+self.length,self.width, **self.kwargs)
        canvas.h_line(self.x_coor,self.y_coor,self.length, **self.kwargs)
        canvas.h_line(self.x_coor+self.width,self.y_coor,self.length, **self.kwargs)

class RightTriangle(Shape):
    def __init__(self,base,height,x_coor,y_coor,max_pairs=16,name="",**kwargs):
        super().__init__(x_coor,y_coor,max_pairs)
        self.base=base
        self.height=height
        self.x_coor=x_coor
        self.y_coor=y_coor
        self.__max_pairs=max_pairs
        self.x2=self.x_coor
        self.y2=self.y_coor+self.height
        self.x3=self.x_coor+self.base
        self.y3=self.y_coor
        self.name=name
        self.kwargs=kwargs
    
    def name(self):
        return self.name
    
    def show_corner(self):
        return Shape.show_x_y(self)

    def show_base(self):
        return self.base

    def show_height(self):
        return self.height

    def area(self,x1=None,y1=None,x2=None,y2=None,x3=None,y3=None):
        if x1==None:
            x1=self.x_coor
        if y1==None:
            y1=self.y_coor
        if x2==None:
            x2=self.x2
        if y2==None:
            y2=self.y2
        if x3==None:
            x3=self.x3
        if y3==None:
            y3=self.y3
        return abs(x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2))/2

    def perimeter(self):
        hypotenuse=(self.base**2+self.height**2)**0.5
        return hypotenuse +self.base +self.height

    def points(self,number,axis):
        if abs(number)<=self.__max_pairs:
            if axis=='Y' or axis=='y':
                range_y=self.height
                increment_y=range_y/abs(number)
                points=[(self.x_coor,self.y_coor)]
                y=self.y_coor
                for i in range(abs(number)):
                    y +=increment_y
                    points.append((self.x_coor,y))
                return points
            elif axis=='X' or axis=='x':
                range_x=self.base
                increment_x=range_x/abs(number)
                points=[(self.x_coor,self.y_coor)]
                x=self.x_coor
                for i in range(abs(number)):
                    x +=increment_x
                    points.append((x,self.y_coor))
                return points
            elif axis=='XY' or axis=='xy':
                range_x=abs(self.base)
                increment_x=range_x/abs(number)
                range_y=abs(self.height)
                increment_y=range_y/abs(number)
                points=[(self.x_coor+self.base,self.y_coor)]
                x=self.x_coor+self.base
                y=self.y_coor
                if self.height>0:
                    if self.base<0:
                        for i in range(abs(number)-1):
                            x +=increment_x
                            y +=increment_y
                            points.append((x,y))
                    elif self.base>0:
                        for i in range(abs(number)-1):
                            x -=increment_x
                            y +=increment_y
                            points.append((x,y))  
                if self.height<0:
                    if self.base<0:
                        for i in range(abs(number)-1):
                            x +=increment_x
                            y -=increment_y
                            points.append((x,y))
                    elif self.base>0:
                        for i in range(abs(number)-1):
                            x -=increment_x
                            y -=increment_y
                            points.append((x,y))
                points.append((self.x_coor,self.y_coor+self.height))
                return points
        else:
            return Shape.base_points(self,number)
    
    def x_y_inside(self,x,y):
        area=self.area()
        area1=self.area(x1=x,y1=y)
        area2=self.area(x2=x,y2=y)
        area3=self.area(x3=x,y3=y)
        return area==area1 + area2 + area3

    def paint(self,canvas):
        canvas.v_line(self.x_coor,self.y_coor+self.base,self.height,**self.kwargs)
        canvas.h_line(self.x_coor+self.height,self.y_coor,self.base,**self.kwargs)
        canvas.line(self.x3,self.y3,self.x2,self.y2,**self.kwargs)  
    
class CompoundShape(Shape):
    def __init__(self,shapes):
        self.shapes=shapes

    def paint(self, canvas):
        for s in self.shapes:
            s.paint(canvas)
            
            
# Initialize objects
box=Rectangle(7,4,10,7,char="#")
right=RightTriangle(8,7,5,10,char="*")
can=Canvas(20,20)
comp=CompoundShape([box,right])

# Draw
comp.paint(can)
can.display()
```

12. Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
class RasterDrawing:
    def __init__(self,canvas_width,canvas_height):
        self.canvas=Canvas(canvas_width,canvas_height)
        self.shapes=dict()
        self.shape_names=list()

    def assign_name(self):
        name_base="shape"
        i=0
        name=name_base+"_"+str(i)
        while name in self.shapes:
            i +=1
            name=name_base+"_"+str(i)
            return name
        return name

    def add_shape(self,shape):
        if shape.name=="":
            shape.name=self.assign_name()
            self.shapes[shape.name]=shape
            self.shape_names.append(shape.name)

    def update(self):
        self.canvas.clear_canvas()
        self.paint(self.canvas)

    def paint(self):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(self.canvas)

    def display(self):
        return self.canvas.display()

    def save(self, filename):
        with open(filename,'w') as file:
            for shape_name in self.shape_names:
                shape=self.shapes[shape_name]
                if isinstance(shape,Rectangle):
                    file.write("\n")
                    file.write("Rectangle("+repr(shape.length)+","+repr(shape.width)+","+repr(shape.x_coor)+","+repr(shape.y_coor)+","+repr(shape.kwargs)+")")
                elif isinstance(shape,RightTriangle):
                    file.write("\n")
                    file.write("RightTriangle("+repr(shape.base)+","+repr(shape.height)+","+repr(shape.x_coor)+","+repr(shape.y_coor)+","+repr(shape.kwargs)+")")

    def load(self, filename):
        with open(filename,'r') as file:
            for line in file:
                line=line.strip()
                if line.startswith("Rectangle("):
                    args=eval(line[len("Rectangle("):-1])
                    shape=Rectangle(*args)
                elif line.startswith("RightTriangle("):
                    args = eval(line[len("RightTriangle("):-1])
                    shape = RightTriangle(*args)
                else:
                    continue
                self.add_shape(shape)
```
